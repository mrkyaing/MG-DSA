```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample POS data
data = {
    'Date': pd.date_range(start='2025-04-01', periods=7, freq='D'),
    'Sales': [200, 340, 560, 430, 710, 300, 620]
}

df = pd.DataFrame(data)

# Top-selling products
top_products = {
    'Product': ['Coffee', 'Sandwich', 'Cake', 'Juice', 'Salad'],
    'Sales': [1200, 950, 850, 600, 450]
}

# Sales by category
categories = ['Beverages', 'Food', 'Desserts']
category_sales = [1800, 2400, 1100]

# Plotting the dashboard with 3 charts
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Simple POS Sales Dashboard', fontsize=16)

# 1. Sales Trend (Line Chart)
axs[0, 0].plot(df['Date'], df['Sales'], marker='o', color='tab:blue')
axs[0, 0].set_title('Sales Trend (Last 7 Days)')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Sales ($)')
axs[0, 0].grid(True)

# 2. Top-Selling Products (Bar Chart)
axs[0, 1].bar(top_products['Product'], top_products['Sales'], color='tab:green')
axs[0, 1].set_title('Top-Selling Products')
axs[0, 1].set_ylabel('Sales ($)')

# 3. Sales by Category (Pie Chart)
axs[1, 0].pie(category_sales, labels=categories, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
axs[1, 0].set_title('Sales by Category')

# Hide the unused subplot (bottom right)
axs[1, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt_path = "D:\\works\\MG-DSA\\pos_sales_dashboard.png"
plt.savefig(plt_path)
plt_path
```
